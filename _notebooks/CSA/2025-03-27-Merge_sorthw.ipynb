# Merge Sort Homework Submission

## Explanation

- **Time Complexity:**  
  - Best, Average, and Worst: **O(n log n)**  
  - This is because the array is divided in half at each step (log n levels) and merging takes linear time.

- **Comparison with Other Sorts:**  
  - **Bubble Sort:** O(n²) (much slower on large arrays)  
  - **Quicksort:** Generally O(n log n) on average but can be O(n²) in the worst-case  
  - **Merge Sort Preference:** Ideal when stable sorting and predictable performance are required.

- **Divide and Conquer:**  
  - Merge sort splits the array into halves, recursively sorts them, and then merges the sorted halves.

- **Stability:**  
  - Merge sort is stable (it keeps equal elements in their original order), which is useful when the order of equal items matters.

## Java Code Submission

```java
import java.util.Arrays;

public class MergeSort {

    // Global counter for comparisons
    private static long comparisonCount = 0;

    // Start merge sort
    public static void mergeSort(int[] arr) {
        if (arr == null || arr.length < 2) return;
        mergeSort(arr, 0, arr.length - 1);
    }

    // Divide the array into halves
    private static void mergeSort(int[] arr, int left, int right) {
        if (left < right) {
            int mid = left + (right - left) / 2;
            mergeSort(arr, left, mid);
            mergeSort(arr, mid + 1, right);
            merge(arr, left, mid, right); // Merge sorted halves
        }
    }

    // Merge two sorted subarrays
    private static void merge(int[] arr, int left, int mid, int right) {
        int n1 = mid - left + 1;
        int n2 = right - mid;
        int[] L = new int[n1];
        int[] R = new int[n2];

        // Copy data into temporary arrays
        for (int i = 0; i < n1; i++) {
            L[i] = arr[left + i];
        }
        for (int j = 0; j < n2; j++) {
            R[j] = arr[mid + 1 + j];
        }

        int i = 0, j = 0, k = left;
        // Merge and count comparisons
        while (i < n1 && j < n2) {
            comparisonCount++;
            if (L[i] <= R[j]) {
                arr[k++] = L[i++];
            } else {
                arr[k++] = R[j++];
            }
        }

        // Copy any remaining elements
        while (i < n1) {
            arr[k++] = L[i++];
        }
        while (j < n2) {
            arr[k++] = R[j++];
        }
    }

    // Main method to test merge sort
    public static void main(String[] args) {
        int[] arr = {38, 27, 43, 3, 9, 82, 10};
        System.out.println("Given Array:");
        System.out.println(Arrays.toString(arr));

        mergeSort(arr);

        System.out.println("\nSorted Array:");
        System.out.println(Arrays.toString(arr));
        System.out.println("Total comparisons made: " + comparisonCount);
    }

    // Overloaded main method for direct testing with int varargs
    public static void main(int... arr) {
        comparisonCount = 0;
        System.out.println("Given Array:");
        System.out.println(Arrays.toString(arr));
        mergeSort(arr);
        System.out.println("Sorted Array:");
        System.out.println(Arrays.toString(arr));
        System.out.println("Total comparisons made: " + comparisonCount);

    }
}

MergeSort.main(38, 27, 43, 3, 9, 82, 10);
```
